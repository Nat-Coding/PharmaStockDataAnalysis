## Japaneese Candle Stick Analysis

To contrast the above visulisation, I have attempted place the pzifer stock over the period of January 2021 to identify the OHLC of the stock during the period after the vaccine has begun to rollout to public population.
As identified in the below article, various countries have begun to recieve vaccinations.

```python
from mpl_finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY

pzifer_reset = pzifer.loc['2020-01':'2020-01'].reset_index()

pzifer_reset['date_ax'] = pzifer_reset['Date'].apply(lambda date: date2num(date))
pzifer_values = [tuple(vals) for vals in pzifer_reset[['date_ax', 'Open', 'High', 'Low', 'Close']].values]

mondays = WeekdayLocator(MONDAY)
alldays = DayLocator()
weekformatter = DateFormatter('%b %d')
dayformatter = DateFormatter('%d')

fig, ax = plt.subplots()
candlestick_ohlc(ax, pzifer_values, width = .8 , colorup='g', colordown='r')
```

## CUMULATIVE RETURN

The below report will address the cumulative return of all four options during this analysis, basically **How much money would I have made for each dollar I invested on 1st January 2020?**

```python
pzifer['Cumulative Return'] = (1+ pzifer['returns']).cumprod()
moderna['Cumulative Return'] = (1+ moderna['returns']).cumprod()
novartis['Cumulative Return'] = (1+ novartis['returns']).cumprod()
csl['Cumulative Return'] = (1+ csl['returns']).cumprod()
```

```python
pzifer.head()
```

```python
pzifer['Cumulative Return'].plot(label='Pzifer',figsize=(15,6))
moderna['Cumulative Return'].plot(label='Moderna')
novartis['Cumulative Return'].plot(label='Novartis')
csl['Cumulative Return'].plot(label='CSL')
plt.legend()
plt.title('Cumulative Return')
plt.ylabel('Stock Growth')
```

**It may be difficult to identify the oppurtunity for investment for any other stock other than Moderna given the above data visulisations. Therefore I have provided a clarified graph to present the three other stock options.**

```python
pzifer['Cumulative Return'].plot(label='Pzifer',figsize=(15,6))
novartis['Cumulative Return'].plot(label='Novartis')
csl['Cumulative Return'].plot(label='CSL')
plt.legend()
plt.title('Cumulative Return')
plt.ylabel('Stock Growth')
```
